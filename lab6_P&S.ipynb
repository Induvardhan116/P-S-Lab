```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 75
std_dev = 10
data_points = np.random.normal(loc=mean, scale=std_dev, size=1000)

plt.hist(data_points, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Data Points')
plt.ylabel('Probability Density')
plt.title('Histogram of Normal Distribution')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.show()

print("Mean:", np.mean(data_points))
print("Standard Deviation:", np.std(data_points))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 20
std_dev = 4

prob_greater_than_23 = 1 - norm.cdf(23, mean, std_dev)

prob_less_than_23 = norm.cdf(23, mean, std_dev)

prob_between_23_35 = norm.cdf(35, mean, std_dev) - norm.cdf(23, mean, std_dev)

x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
y = norm.pdf(x, mean, std_dev)

plt.plot(x, y)
plt.fill_between(x[x >= 23], 0, norm.pdf(x[x >= 23], mean, std_dev), alpha=0.5, color='red')
plt.fill_between(x[x <= 23], 0, norm.pdf(x[x <= 23], mean, std_dev), alpha=0.5, color='green')
plt.fill_between(x[(x >= 23) & (x <= 35)], 0, norm.pdf(x[(x >= 23) & (x <= 35)], mean, std_dev), alpha=0.5, color='blue')

plt.xlabel('BMI')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of BMI')

plt.text(25, 0.05, f'P(BMI >= 23) = {prob_greater_than_23:.3f}', color='red')
plt.text(15, 0.05, f'P(BMI <= 23) = {prob_less_than_23:.3f}', color='green')
plt.text(29, 0.02, f'P(23 <= BMI <= 35) = {prob_between_23_35:.3f}', color='blue')

plt.show()

print("Probability of BMI greater than or equal to 23:", prob_greater_than_23)
print("Probability of BMI less than or equal to 23:", prob_less_than_23)
print("Probability of BMI between 23 and 35:", prob_between_23_35)
```
